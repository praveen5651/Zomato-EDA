```python
highest_order = final_df['Country'].value_counts()
country_names = final_df.Country.value_counts().index
```

# Pie Chart 


```python
plt.figure(figsize=(10,10))
plt.pie(highest_order[:5], labels=country_names[:5], autopct="%2f%%")
```

* Highest sales are happend in india compare to United States, United Kingdom, Brazil and UAE
* India is having 93.1725% 

# Rating Type Analysis 

```python
# Grouping this related columns 
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text','Votes']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x='Rating color',y='Rating Count', data=ratings)
```

# Observations
* No rating count is very high.
* it is denoted by white

```python
plt.figure(figsize=(50,200))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color', data=ratings)
```

# Observations
* Zero ratings are high
* Average rating between 2.5 to 3.4

# Count Plot

```python
sns.countplot(x='Rating color',data=ratings)
```

```python
sns.countplot(x='Rating text',data=ratings)
```
